# [創作課題]PID制御とゲインK

## 問題
旋回軸を支点として回転するアームを目標とする角度で静止させる事を考える.

## 方法
[Pythonによる制御工学入門5章](https://www.ohmsha.co.jp/book/9784274223907/)のコードを参考として主にPythonを用いて制御系をシュミレートし、システムの出力について考察する.

## 概要
旋回軸を支点として回転するアームを目標とする角度で静止させる制御を例として、
最も一般的なフィードバック制御であるPID制御とゲインとして与えられる各パラメータの関係について考える.

## 制御対象となる系のモデル化
慣性モーメントを$J$、アームの質量を$M$、重力加速度を$g$と旋回軸に作用する摩擦力を$\mu$、原点からアームの重心の距離を$l$とする.<br />
運動エネルギー$K = J\displaystyle\frac{d^{2}\theta}{dt^{2}} + \mu\displaystyle\frac{d\theta}{dt}$<br/>
位置エネルギー$P = -mgl\sin(\theta)$<br/>
ラグランジュ関数$L = P - K$へ値を代入.<br />
$L = J\displaystyle\frac{d^{2}\theta}{dt^{2}} + \mu\displaystyle\frac{d\theta}{dt} + mgl\sin(\theta)$<br />
$\theta = 0$近傍で$\theta(t) = y(t)$と近似する.<br />
$L = J\displaystyle\frac{d^{2}y(t)}{dt^{2}} + \mu\displaystyle\frac{dy(t)}{dt} + mgl y(t)$<br />
$L$は系に作用する物理的エネルギーの総和なので$u(t) = L$とすると<br />
$u(t) = J\displaystyle\frac{d^{2}y(t)}{dt^{2}} + \mu\displaystyle\frac{dy(t)}{dt} + mgl y(t)$<br />
両辺をラプラス変換する.<br />
$u(s) = Js^{2}y(s) + \mu s y(s) + mgl y(s)$<br />
$\displaystyle\frac{y(s)}{u(s)} = \displaystyle\frac{1}{J s^{2} + \mu s + mgl}$<br>
$P(s) = \displaystyle\frac{1}{J s^{2} + \mu s + mgl} \cdots (1 - 1)$


## 閉ループ系における制御の論理
![画像が表示されなかった時に表示される文字列](img\close_loop_001.png)

### 一般化された伝達関数の導出
$G(s) = \displaystyle\frac{K(s)P(s)}{1 + K(s)P(s)}$<br />
$y(s) = P(s)(u(s) + d(s))$<br />
$y(s) = P(s)u(s) + P(s)d(s) \cdots (1)$<br />
$u(s) = (r(s) - y(s))K(s) \cdots (2)$<br />
$(1)$へ$(2)$を代入.
$y(s) = P(s)K(s)(r(s) - y(s)) + P(s)d(s)$
$(1 + P(s)K(s))y(s) = P(s)K(s)r(s) + P(s)d(s)$
$y(s) = \displaystyle\frac{P(s)K(s)}{1 + P(s)K(s)}r(s) + \displaystyle\frac{P(s)}{1 + P(s)K(s)}d(s) \cdots (3)$<br />
以上より$G_{yr}(s) = \displaystyle\frac{P(s)K(s)}{1 + P(s)K(s)}, G_{yd}(s) = \displaystyle\frac{P(s)}{1 + P(s)K(s)} \cdots (1 - 2)$
を得る.<br />

$(2)$より<br />
$u(s) = K(s)(r(s) - y(s))$<br />
$y(s) = P(s)(u(s) + d(s)) \cdots (5)$
$(2)$へ$(5)$を代入.<br />
$u(s) = K(s)r(s) - P(s)K(s)(u(s) + d(s))$<br />
$(1 + P(s)K(s))u(s) = P(s)K(s)d(s)$<br />
$u(s) = \displaystyle\frac{K(s)}{1 + P(s)K(s)}r(s) + \displaystyle\frac{P(s)K(s)}{1 + P(s)K(s)}d(s)$<br />
以上より$G_{ur}(s) = \displaystyle\frac{K(s)}{1 + P(s)K(s)}, G_{ud}(s) \displaystyle\frac{P(s)}{1 + P(s)K(s)} \cdots (1 - 3)$を得る.<br />
閉ループ系が安定なフィードバック制御となる条件は$(1 - 1)$および$(1 - 2)$式の入力/出力がすべて安定であり内部安定性を保障することとなる.<br />


### PIDコントローラの数理的模式化
コントローラの操作量を$u(t)$とするとPIDコントローラは、比例制御器＋積分器＋微分器の多項式として、以下の数式で表現できる.<br />
$
u(t) = \it{K}_{P}e(t) + \it{k}_{I}\displaystyle\int_{t}^{0} e(\tau) d\tau + \it{k}_{D}e(t)
$<br />
上式をラプラス変換すると<br />
$u(s) = \it{k}_{P}e(s) + \displaystyle\frac{\it{k}_{I}}{s}e(s) + \it{k}_{D}s e(s) $を得る.<br />

### PIDコントローラを組み込んだ系のブロック線図
$\ast$外乱を$d$出力項を$y$とする<br />
![画像が表示されなかった時に表示される文字列](img\PID_001.png)

### PIDコントローラの操作量
#### $\it{k}_{P}$ 比例動作(P動作)
- $制御量y(t)目標値r(t)の偏差e(t) := r(t) - y(t)が大きければシステムに対する入力値u(t)を大きくする$
- $制御量y(t)目標値r(t)の偏差e(t) := r(t) - y(t)が小さければシステムに対する入力値u(t)を小さくする$

$\ast 現在時刻の情報を入力値へ反映する$

#### $\it{k}_{I}$積分動作(I動作)
仮に比例動作のみで制御を行う場合操作量$u(t)$と静止摩擦等の反力が釣り合い定常偏差を生じる.<br />
偏差$e(t)$を積分することで定常偏差を定量的に把握し制御に反映するする事が出来る.<br />

#### $\it{k}_{D}$微分動作(D動作)
$\it{k}_{P}$や$\it{k}_{I}$の値を適正な範囲に留めてオーバーシュートを抑止するため、偏差$e(t)$の微分値を利用する.

## python並びにmatlabのcontroll.bode関数に関する補足
python並びにmatlabのbode関数の出力地は0近傍で$360^{\circ} \times n (n \in N)$倍の誤差を生じることがある.<br />
上記に対応するため以下の対策を行う.<br />
- ボード線図を出力するメソッド内で位相の正規化を行うことがある
- bode関数の誤差に対応するため可能な場合はpythonのコードで描画すべき位相線図をmatlabで描画された位相線図に差し替えることがある
- matlabでpython同様に位相線図の描画が困難な場合、位相線図を提示せず周波数特性への言及のみを記載することがある

[bode関数で描いた​ボード線図について](https://jp.mathworks.com/matlabcentral/answers/103525-bode-0)

```python
import cupy as cp
import math
import numpy as np
import pandas as pd
import sympy as sp
import matplotlib.pyplot as plt
import control

from control.matlab import *
from scipy.integrate import odeint

from Modules import Modules


%matplotlib inline

print("OK\n")
```

```python
modules = Modules('')

plt.rcParams['font.family'] ='sans-serif'
plt.rcParams['xtick.direction'] = 'in'
plt.rcParams['ytick.direction'] = 'in'
plt.rcParams['xtick.major.width'] = 1.0
plt.rcParams['ytick.major.width'] = 1.0
plt.rcParams['font.size'] = 10
plt.rcParams['axes.linewidth'] = 1.0
plt.rcParams['mathtext.default'] = 'regular'
plt.rcParams['axes.xmargin'] = '0'
plt.rcParams['axes.ymargin'] = '0.05'
plt.rcParams['savefig.facecolor'] = 'None'
plt.rcParams['savefig.edgecolor'] = 'None'
```

```python
g = 9.8
l = 0.2
M = 0.5
mu = 1.5e-2
J = 1.0e-2

P = tf([0, 1], [J, mu, M*g*l])

ref = 30
```

## 閉ループ系におけるフィードバック制御をおこなわない場合のアーム系の応答
![画像が表示されなかった時に表示される文字列](img\closed_loop.png)

```python
def Control(max):
    try:
        LS = modules.linestyle_generator()
        fig, ax = plt.subplots()
        Gyr = feedback(P , 1)
        y,t = step(Gyr, np.arange(0, max, 0.01))

        pltargs = {'ls': next(LS), 'label': '$P$'}
        ax.plot(t, y * ref, **pltargs)
        ax.axhline(ref, color="k", linewidth = 0.5)
        modules.plot_set(ax, 't', 'y', 'best')
    except NameError as err:
        print("NameError: {0}".format(err))
    except TypeError as err:
        print("TypeError: {0}".format(err))
    except ValueError as err:
        print("ValueError: {0}".format(err))
    except OSError as err:
        print("OS error: {0}".format(err))
    except BaseException as err:
        print(f"Unexpected {err=}, {type(err)=}")
        raise
```

```python
def bord(param):
    try:
        LS = modules.linestyle_generator()
        fig, ax = plt.subplots(2, 1)
        
        Gyr = feedback(P, 1)
        gain, phase, w = bode(Gyr, logspace(-1, param), plot=False)
        
        pltargs = {'ls': next(LS), 'label': '$P$'}
        ax[0].semilogx(w, 20 * np.log10(gain), **pltargs)
        ax[1].semilogx(w, (phase * (180/np.pi)), **pltargs)
        modules.bodeplot_set(ax, 'lower left')
    except NameError as err:
        print("NameError: {0}".format(err))
    except TypeError as err:
        print("TypeError: {0}".format(err))
    except ValueError as err:
        print("ValueError: {0}".format(err))
    except OSError as err:
        print("OS error: {0}".format(err))
    except BaseException as err:
        print(f"Unexpected {err=}, {type(err)=}")
        raise
```

#### 制御項として比例成分のみを印可した場合に関する考察
- $\it{k}_{P}$の値が十分に小さい場合は行き過ぎ量の収束は遅く定常偏差の目標値からの乖離は大きい
- $\it{k}_{P}$の値が十分に大きい場合は行き過ぎ量の収束は早く定常偏差の目標値からの乖離は小さい
- 系に印可される制御量が$\it{k}_{P}$のみの場合は正負の行き過ぎ量は完全に目標値へ収束せず振動を続ける
- $\it{k}_{P}$の値を大きくするとバンド幅が広がり応答は早くなる
- $\it{k}_{P}$の値を大きくするとゲイン線図がグラフの上方向へ移動し低周波成分の作用でバンド幅が大きくなる
以上より系に印可される$\it{k}_{P}$の値を大きくすることで応答速度が速くなる一方、系の出力は振動的となる.
また上記実験は$\it{k}_{P}$の値に制限はないものとして実施したが、実際のコントローラ設計においては系の許容できる範囲でパラメータの上限値が定まるため、比例制御のみでアームの角度を制御することは原則として不可能である。

### PD制御による比例ゲインと系の応答
系に対して制御項として比例成分に加えて微分成分を印可した場合を考える.

![画像が表示されなかった時に表示される文字列](img\PD_001.png)

```python
def PD_Control(kp, kd, max):
    try:
        LS = modules.linestyle_generator()
        fig, ax = plt.subplots()
        for i in range(len(kd)):
            K = tf([kd[i], kp], [0, 1])
            Gyr = feedback(P * K, 1)
            y,t = step(Gyr, np.arange(0, max, 0.01))

            pltargs = {'ls': next(LS), 'label': '$K_D$='+str(kd[i])}
            ax.plot(t, y * ref, **pltargs)
        ax.axhline(ref, color="k", linewidth = 0.5)
        modules.plot_set(ax, 't', 'y', 'best')
    except NameError as err:
        print("NameError: {0}".format(err))
    except TypeError as err:
        print("TypeError: {0}".format(err))
    except ValueError as err:
        print("ValueError: {0}".format(err))
    except OSError as err:
        print("OS error: {0}".format(err))
    except BaseException as err:
        print(f"Unexpected {err=}, {type(err)=}")
        raise
```

```python
def bord_PD(kp, kd, param, swich):
    try:
        LS = modules.linestyle_generator()
        fig, ax = plt.subplots(2, 1) 

        for i in range(len(kd)):
            K = tf([kd[i], kp], [0, 1])
            Gyr = feedback(P * K, 1)
            y,t = step(Gyr,np.arange(0, param, pow(10, -4)))
            gain, phase, w = bode(Gyr, logspace(-1, param), plot=False)
            
            # 計算結果が遅角するため補正する
            array = phase.tolist()
            if i in swich:
                for j, x in enumerate(array):
                    array[j] = x + 6.28319
            phase = np.array(array)
            
            pltargs = {'ls': next(LS), 'label': '$k_D$='+str(kd[i])} 
            ax[0].semilogx(w, 20*np.log10(gain), **pltargs)
            ax[1].semilogx(w, np.rad2deg(phase), **pltargs)
        modules.bodeplot_set(ax, 'best')
        ax[1].set_ylim(-190,10)
        ax[1].set_yticks([-180,-90,0])
        fig.tight_layout()
    except NameError as err:
        print("NameError: {0}".format(err))
    except TypeError as err:
        print("TypeError: {0}".format(err))
    except ValueError as err:
        print("ValueError: {0}".format(err))
    except OSError as err:
        print("OS error: {0}".format(err))
    except BaseException as err:
        print(f"Unexpected {err=}, {type(err)=}")
        raise
```

#### 制御項として比例成分と微分成分を印可した場合に関する考察
- 比例成分$k_{P}$のゲインが大きすぎず適正な値である場合、場合微分成分$k_{D}$のゲインを大きな値とすることで振動を抑制できる
- 比例成分$k_{P}$並びに微分成分のみで定常偏差を0に近付けることはできない
- 比例成分$k_{P}$のゲインを大きくする事で定常偏差の絶対値は縮小するが応答は振動的となる
以上より制御系に微分成分を加えることで振動を抑制することはできるが定常偏差を0に近付けることは出来ない.

### PI制御による比例ゲインと系の応答
系に対して制御項として比例成分に加えて微分成分を印可した場合を考える.
#![画像が表示されなかった時に表示される文字列](img\PI_001.png)

```python
def PI_Control(kp, ki, max):
    try:
        LS = modules.linestyle_generator()
        fig, ax = plt.subplots()
        for i in range(len(ki)):
            K = tf([kp, ki[i]], [1, 0])
            Gyr = feedback(P * K, 1)
            y,t = step(Gyr, np.arange(0, max, 0.01))

            pltargs = {'ls': next(LS), 'label': '$K_I$='+str(ki[i])}
            ax.plot(t, y * ref, **pltargs)
        ax.axhline(ref, color="k", linewidth = 0.5)
        modules.plot_set(ax, 't', 'y', 'best')
    except NameError as err:
        print("NameError: {0}".format(err))
    except TypeError as err:
        print("TypeError: {0}".format(err))
    except ValueError as err:
        print("ValueError: {0}".format(err))
    except OSError as err:
        print("OS error: {0}".format(err))
    except BaseException as err:
        print(f"Unexpected {err=}, {type(err)=}")
        raise
```

```python
def bord_PI(kp, ki, param, swich):
    try:
        LS = modules.linestyle_generator()
        fig, ax = plt.subplots(2, 1) 

        for i in range(len(ki)):
            K = tf([kp, ki[i]], [1, 0])
            Gyr = feedback(P * K, 1)
            gain, phase, w = bode(Gyr, logspace(-1, param), plot=False)
            
            # 計算結果が反転するため補正する
            array = phase.tolist()
            if i in swich:
                for j, x in enumerate(array):
                    array[j] = (x - 6.28319)
            phase = np.array(array)

            pltargs = {'ls': next(LS), 'label': '$k_I$='+str(ki[i])} 
            ax[0].semilogx(w, 20*np.log10(gain), **pltargs)
            ax[1].semilogx(w, np.rad2deg(phase), **pltargs)
        modules.bodeplot_set(ax, 'best')
        ax[1].set_ylim(-190,0)
        ax[1].set_yticks([-180,-90,0])
        fig.tight_layout()
    except NameError as err:
        print("NameError: {0}".format(err))
    except TypeError as err:
        print("TypeError: {0}".format(err))
    except ValueError as err:
        print("ValueError: {0}".format(err))
    except OSError as err:
        print("OS error: {0}".format(err))
    except BaseException as err:
        print(f"Unexpected {err=}, {type(err)=}")
        raise
```

#### 制御項として比例成分と積分成分を印可した場合に関する考察
- 比例成分$k_{P}$に加えて積分成分$k_{I}$を印可することで定常偏差の値を0に近付けることができる
- 比例成分$k_{P}$に加えて積分成分$k_{I}$を印可しても応答は振動的となる
- 積分成分$k_{I}$のゲインを大きくすることで振動が増幅され応答は発散する
比例成分を一定の値として積分成分のゲインを大きな値にした場合、定常偏差の縮小と振動がトレードオフの関係となる<br />.
以上より、比例成分と積分成分のみで制御器を構成することは系全体の実用性という観点で好ましくないと考えられる.

## モデルマッチングによる最適ゲイン導出
PI-D制御を行う系に対してモデルマッチングにより最適ゲインを導出する.

### PI-D制御の模式化
![画像が表示されなかった時に表示される文字列](img\PI_D_001.png)

$u(t) = K_{I}\displaystyle\int_{0}^{\tau}e(t)dt + k_{P}e(t) + K_{d}y(t)$<br />
両辺ラプラス変換.<br />
$u(s) = K_{P}e(s) + \displaystyle\frac{K_{I}}{s}e(s) + K_{I}s y(s) \cdots (1)$<br />
$e(s) = r(s) - y(s) \cdots (2)$
$(1)$へ$(2)$を代入.<br />
$u(s) = K_{P}(r(s) - y(s)) + \displaystyle\frac{K_{I}}{s}(r(s) - y(s)) - K_{D}y(s)$<br />
$u(s) = \displaystyle\frac{K_{D}s^{2} + K_{P}s + K_{I}}{s} \displaystyle\left( \displaystyle\frac{K_{D}s + K_{I}}{K_{D}s^{2} + K_{P}s + K_{I}}r(s) - y(s) \right) \cdots (3)$<br />
$(3)$右辺の係数に着目して<br />
$K_{1}(s) = \displaystyle\frac{K_{D}s^{2} + K_{P}s + K_{I}}{s}, K_{2}(s) = \displaystyle\frac{K_{D}s + K_{I}}{K_{D}s^{2} + K_{P}s + K_{I}} \cdots (3)$を得る.<br />
![画像が表示されなかった時に表示される文字列](img\PI_D_002.png)

$G_{yr} = K_{2}(s)\displaystyle\frac{K_{1}(s)K_{P}(s)}{1 + K_{1}(s)K_{P}(s)} \cdots (4)$<br />
$(4)$へ$(3)$を代入.<br />
$G_{yr}(S) = \displaystyle\frac{K_{P}s + K_{I}}{ Js^{3} + (\mu + K_{D})s^{2} + (Mgl + K_{P})s + K_{I} } \cdots (5)$
以上より伝達関数$G_{yr}(S)$を得る.<br />

### モデルマッチングの定義
模範モデル$M(s) = \displaystyle\frac{\omega_{n}^{2}}{ s^{2} + 2\zeta\omega_{n}s + \omega_{n}^{2} } \cdots (6)$とする.<br />

$(5)$、$(6)$より模範モデルを$M(s)$、伝達関数を$G_{yr}(s)$とすると、$\displaystyle\frac{1}{G_{yr}(s)}$と$\displaystyle\frac{1}{M(s)}$のマクローリン展開を導出し、3次多項式の各項の係数を比較し制御パラメータ$K_{I}$、$K_{P}$、$K_{D}$を導出する.


### $\displaystyle\frac{1}{G_{yr}}$のマクローリン展開

$\displaystyle\frac{1}{G_{yr}} = 
{\left(\frac{J}{\mathrm{K_{I}}}-\frac{{\mathrm{K_{P}}}^3 }{{\mathrm{K_{I}}}^3 }+\frac{{\mathrm{K_{P}}}^2 \,{\left(\mathrm{Mgl}+\mathrm{K_{P}}\right)}}{{\mathrm{K_{I}}}^3 }-\frac{\mathrm{K_{P}}\,{\left(\mathrm{K_{D}}+\mu \right)}}{{\mathrm{K_{I}}}^2 }\right)}\,s^3 +{\left(\frac{\mathrm{K_{D}}+\mu }{\mathrm{K_{I}}}+\frac{{\mathrm{K_{P}}}^2 }{{\mathrm{K_{I}}}^2 }-\frac{\mathrm{K_{P}}\,{\left(\mathrm{Mgl}+\mathrm{K_{P}}\right)}}{{\mathrm{K_{I}}}^2 }\right)}\,s^2 +{\left(\frac{\mathrm{Mgl}+\mathrm{K_{P}}}{\mathrm{K_{I}}}-\frac{\mathrm{K_{P}}}{\mathrm{K_{I}}}\right)}\,s+1$
<br />

### $\displaystyle\frac{1}{M(s)}$のマクローリン展開
$\displaystyle\frac{1}{M(s)} = 
1 + \displaystyle\frac{2\zeta}{\omega_{n}}s + \displaystyle\frac{1}{\omega_{n}}s^{2}
$



### 模範モデルと$G_{yr}$の項を比較

$k_{p} = J\,{\mathrm{wn}}^2$
<br />

$k_{I} = \displaystyle\frac{\mathrm{Mgl}\,\mathrm{wn}}{2\,\zeta}$
<br />

$k_{D} = 2\zeta\omega_{n}J + \displaystyle\frac{Mgl}{2\zeta\omega_{n}} - \mu$

### モデルマッチングによる最適ゲインの導出

```python
g  = 9.81
l  = 0.2
M  = 0.5
mu = 1.5e-2
J  = 1.0e-2

P = tf( [0,1], [J, mu, M*g*l] )

ref = 30
```

```python
def model1(param = None, omega = None, max = 2):
    try:
        zeta = None
        omega_n = 15
        
        if param is None:
            zeta = (1, 1/np.sqrt(2))
        else:
            zeta = (1, param)
            
        if omega is None:
            omega_n = 15
        else:
            omega_n = omega
   
        Label = ('Binomial coeff.', 'Butterworth')

        LS = modules.linestyle_generator()
        fig, ax = plt.subplots()

        for i in range(2):
            Msys = tf([0,omega_n**2], [1,2*zeta[i]*omega_n,omega_n**2])
            y, t = step(Msys, np.arange(0, max, 0.01))

            ax.plot(t, y*ref, ls=next(LS), label=Label[i])

        ax.set_xlim(0, max)
        ax.set_ylim(0, 50)
        modules.plot_set(ax, 't', 'y', 'best')
    except NameError as err:
        print("NameError: {0}".format(err))
    except TypeError as err:
        print("TypeError: {0}".format(err))
    except ValueError as err:
        print("ValueError: {0}".format(err))
    except OSError as err:
        print("OS error: {0}".format(err))
    except BaseException as err:
        print(f"Unexpected {err=}, {type(err)=}")
        raise  
```

```python
def matching(z = None, omega = None, max = 2):
    try:
        zeta = None
        omega_n = None

        if z is None:
            zeta = 1/np.sqrt(2)
        else:
            zeta = z

        if omega is None:
            omega_n = 15
        else:
            omega_n = omega        

        Msys = tf([0, pow(omega_n, 2)], [1, 2 * zeta * omega_n, pow(omega_n, 2)])
        
        kp = pow(omega_n, 2) * J
        ki = (omega_n * M * g * l)/(2 * zeta)
        kd = (2 * zeta * omega_n * J) + (M * g * l)/(2 * zeta * omega_n) - mu

        print('kP=', kp)
        print('kI=', ki)
        print('kD=', kd)

        Gyr = tf([kp, ki], [J, mu + kd, M * g * l+ kp, ki])

        yM, tM = step(Msys, np.arange(0, max, 0.01))
        y, t = step(Gyr, np.arange(0, max, 0.01))

        fig, ax = plt.subplots()

        ax.plot(tM, yM * ref, label='M', ls = '-.')
        ax.plot(t, y * ref, label='Gyr')

        ax.set_xlim(0, max)
        ax.set_ylim(0, 50)
        modules.plot_set(ax, 't', 'y', 'best')
    except NameError as err:
        print("NameError: {0}".format(err))
    except TypeError as err:
        print("TypeError: {0}".format(err))
    except ValueError as err:
        print("ValueError: {0}".format(err))
    except OSError as err:
        print("OS error: {0}".format(err))
    except BaseException as err:
        print(f"Unexpected {err=}, {type(err)=}")
        raise  
```

#### 模範モデルの応答

```python
model1(1/np.sqrt(2), 15, 2)
```

#### モデルマッチングによる制御ゲインの導出と応答

```python
matching(1/np.sqrt(2), 15, 2)
```

## 課題のまとめ
本課題では最適な制御パラメータを得るため、系に対してPI-D制御を適用し制御ゲインの導出にモデルマッチングを用いた.<br/>
結果、振動を最小限に抑制しつつ、短時間で目標の角度へ収束する制御を実現できた.<br/>
他方で本課題では比較的単純な1軸の系に対して制御を適用したが、制御対象となる系が複雑化した場合については、サブシステム間の相互作用を反映したモデルを構築するノウハウが必要になると考える.
